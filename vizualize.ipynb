```python
import json
import matplotlib.pyplot as plt
from record_utils import RecordManager 

def plot_metrics(metrics1, metrics2, label1, label2):
    """
    Plot train accuracy, test accuracy, and loss between two datasets.

    Parameters:
    - metrics1: Metrics from the first file (dict)
    - metrics2: Metrics from the second file (dict)
    - label1: Label for the first file's data
    - label2: Label for the second file's data
    """
    epochs = metrics1["epochs"]  # Assume both files have the same number of epochs

    # Plot Train Accuracy
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, metrics1["train_accuracies"], label=f"{label1} - Train Accuracy", linestyle="--", marker="o")
    plt.plot(epochs, metrics2["train_accuracies"], label=f"{label2} - Train Accuracy", linestyle="--", marker="x")
    plt.title("Train Accuracy Comparison")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot Test Accuracy
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, metrics1["test_accuracies"], label=f"{label1} - Test Accuracy", linestyle="-", marker="o")
    plt.plot(epochs, metrics2["test_accuracies"], label=f"{label2} - Test Accuracy", linestyle="-", marker="x")
    plt.title("Test Accuracy Comparison")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot Loss
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, metrics1["losses"], label=f"{label1} - Loss", linestyle=":", marker="o")
    plt.plot(epochs, metrics2["losses"], label=f"{label2} - Loss", linestyle=":", marker="x")
    plt.title("Loss Comparison")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
file_path1 = "models/cifar10/run3/accuracies/all_accuracies.json"
file_path2 = "models/cifar10_pure_adversarial/run6/accuracies/all_accuracies.json"
# Load metrics from both files
metrics1 = RecordManager.load_metrics(file_path1)
metrics2 = RecordManager.load_metrics(file_path2)
# Plot and compare metrics
plot_metrics(metrics1, metrics2, label1="Clean", label2="Adversarial")
```
